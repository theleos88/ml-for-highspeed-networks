```python
data = pd.read_csv('./data/fulldataset.csv')
data
```

```python
X = data.iloc[:,4:].values
X = (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
tsne = manifold.TSNE(n_components=2, perplexity=500, n_iter=1000)
X = tsne.fit_transform(X)
```
